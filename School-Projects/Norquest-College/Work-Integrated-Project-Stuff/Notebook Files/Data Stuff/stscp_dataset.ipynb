```python
# Obtaining the "Project List" DataFrame Length to Use for Null Ratio % Calculations (Used as Denominator)...

df_len = df_plist.shape[0]

# Obtaining Column Names (Keys) for DataFrame Access in-loop for Null Ratio % Calculations for "Project List" DataFrame...

df_col_keys = list(df_plist.columns)   # Returned Index Object is Converted to List Type


# Initializing an Empty List for Row-Creation in the Null Result DataFrame...
rows = []

# Iterating through "Project List" DataFrame Keys for Null % Calculations

for key in df_col_keys:

  null_count = df_plist[key].isnull().sum()   # Iterated Key to Access Project List DataFrame for Null Value Count (Sum)

  null_ratio = (null_count / df_len) * 100    # Ratio is Taken (How Many Null Values in the Column VS The Total # of Rows)

  result_tup = (key, null_ratio)              # Key and Calculated Result Stored as Tuple
  rows.append(result_tup)                     # Result Tuple Appended to Rows List


# Creating a Results DataFrame to Nicely Display the % of Nulls per Column

results_df = pd.DataFrame(rows, columns=["Column Name", "% of Nulls"])
```

```python
# Cross-Tabulation between Co-Operatives & State

cross_tab = pd.crosstab(df_plist['State'], df_plist['Utility Type'])

# Focus strictly on Cooperatives to see the "correlation" to geography
coop_by_state = cross_tab['Cooperative'].sort_values(ascending=False)
print(coop_by_state)
```

```python
# Importing Chi-Squared Testing Module
from scipy.stats import chi2_contingency

# Establishing Ch-Squared Test Variables
chi2, p, dof, ex = chi2_contingency(cross_tab)

# Displaying Chi-Squared Test Results between State & Utility-Type
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")                  # If p < 0.05, there is a strong "correlation" between state and type
```

**EDA - DATA VISUALIZATION**

```python

#The sum of the system size for states
state_map_data = df_plist.groupby('State')['System Size (MW-AC)'].sum().reset_index()

#Generate the Choropleth Map
fig = px.choropleth(
    state_map_data,
    locations='State',           #States column
    locationmode="USA-states",   #tells Plotly to use a US map
    color='System Size (MW-AC)',
    scope="usa",                 #limit the map to the USA
    color_continuous_scale="Viridis",
    title='Total Community Solar Capacity (MW-AC) by State',
    labels={'System Size (MW-AC)': 'Total MW-AC'}
)

fig.show()
```

```python
# Filtering DataFrame for 'Cooperative' Utility Types...
df_coops = df_plist[df_plist['Utility Type'] == 'Cooperative'].copy()

# Grouping System Sizes in Cooperative utility Projects by State
state_coop_data = df_coops.groupby('State')['System Size (MW-AC)'].sum().reset_index()

#Generate the Choropleth Map
fig = px.choropleth(
    state_coop_data,
    locations='State',           #States column
    locationmode="USA-states",   #tells Plotly to use a US map
    color='System Size (MW-AC)',
    scope="usa",                 #limit the map to the USA
    color_continuous_scale="Viridis",
    title='Total Cooperative Solar Capacity (MW-AC) by State',
    labels={'System Size (MW-AC)': 'Total MW-AC'}
)

fig.show()
```

```python
#Group by 'Year of Interconnection' and count the System Size per Year
yearly_coop_output = df_coops.groupby('Year of Interconnection')['System Size (MW-AC)'].count()

#plot the line chart
plt.figure(figsize=(12, 6))
yearly_coop_output.plot(kind='line', marker='o', color='teal', linewidth=2)

#adding titles and labels
plt.title('Cooperative Solar Energy System Sizes per Year', fontsize=14)
plt.xlabel('Year of Interconnection', fontsize=12)
plt.ylabel('Size of Solar Energy Output', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python

yearly_all_output = df_plist.groupby("Year of Interconnection")['System Size (MW-AC)'].count()

#plot the line chart
plt.figure(figsize=(12, 6))
yearly_all_output.plot(kind='line', marker='o', color='salmon', linewidth=2)

#adding titles and labels
plt.title('All-Types: Solar Energy System Sizes per Year', fontsize=14)
plt.xlabel('Year of Interconnection', fontsize=12)
plt.ylabel('Size of Solar Energy Output', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
#Group by both Year and Utility Type, then count the records
utility_trends = df_plist.groupby(['Year of Interconnection', 'Utility Type']).size().unstack(fill_value=0)

#plotting the multiv line chart
#'marker' adds dots to each year
ax = utility_trends.plot(kind='line', marker='o', figsize=(14, 8), linewidth=2)

plt.title('Annual Project Trends by Utility Type', fontsize=16)
plt.ylabel('Number of New Projects', fontsize=12)
plt.xlabel('Year of Interconnection', fontsize=12)

#moving the legend outside the plot so it doesn't cover the lines
plt.legend(title='Utility Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Feature Engineering Section**

```python
# Creating New Feature - Solar Efficiency (System AC / System DC)

df_plist["acdc_ratio_kW"] = (
    df_plist["System Size (kW-AC)"] /
    df_plist["System Size (kW-DC)"]
)

""" Essentially indicates system design and inverter efficiency/sizing of panel capacity.
Values ~1.0 indicate power-inverter is sized to panel capacity nearly perfectly.

Values < 1.0 indicate panels are over-sized compared to power-inverter. Common in modern design.
Great for low-light conditions.

Can Correlate with Project Size, Utility Type, Program Designs... """

print() # This line is only here to clear the terminal output from the above comment
```

```python
# Creating New Feature - Project Age

# Getting the Latest Year for Interconnection in "Project List" DataFrame
latest_year = df_plist['Year of Interconnection'].max()

# Subtracting Latest Year from Interconnection Year, storing result in 'Project Age'

df_plist['Project Age'] = latest_year - df_plist['Year of Interconnection']
```
